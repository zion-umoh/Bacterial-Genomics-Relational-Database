## Convert GAF file into a CSV file

```python
import os
from pathlib import Path

# Define the input directory
input_files = 'input_files'

# Get all gaf files
gaf_files = [f for f in os.listdir(input_files) if f.endswith('.gaf')]
print(f"Found {len(gaf_files)} GAF file(s):")
for file in gaf_files:
    print(f"  - {file}")
```

```python
import pandas as pd

for gaf_file in gaf_files:
    gaf_path = Path(input_files) / gaf_file
    df = pd.read_csv(gaf_path, sep='\t', comment='!', header=None, dtype=str)
    csv_path = gaf_path.with_suffix('.csv')
    df.to_csv(csv_path, index=False)
    print(f"Converted {gaf_file} to {csv_path.name}")
    display(df.head())
```

```python
column_names = [
    "DB",
    "DB Object ID",
    "DB Object Symbol",
    "Qualifier",
    "GO ID",
    "DB:Reference",
    "Evidence Code",
    "With (or) From",
    "Aspect",
    "DB Object Name",
    "DB Object Synonym",
    "DB Object Type",
    "Taxon",
    "Date",
    "Assigned By",
    "Annotation Extension",
    "Gene Product Form ID",
]

all_dfs = []
for gaf_file in gaf_files:
    gaf_path = Path(input_files) / gaf_file
    temp_df = pd.read_csv(gaf_path, sep="\t", comment="!", header=None, dtype=str)
    temp_df.columns = column_names
    all_dfs.append(temp_df)

combined_df = pd.concat(all_dfs, ignore_index=True)
combined_csv_path = Path(input_files) / "combined_annotations.csv"
combined_df.to_csv(combined_csv_path, index=False)

print(f"Saved combined file to {combined_csv_path}")
display(combined_df.head())
```
